Problem 1

```python
def gradplot(x1, x2, lr, gm, optval=-1):
  prev=1000000
  cur=1
  ct=1
  ep=[]
  err=[]
  while (1):
    x1=x1-lr*(2*x1-x2-1)
    x2=x2-lr*(2*gm*x2-x1-1)
    cur=x1*x1+gm*x2*x2-x1*x2-x1-x2
    ep.append(ct)
    err.append(abs(optval-cur))
    if(cur==prev):
      optx1=x1
      optx2=x2
      optf=cur
      break
    prev=cur
    ct+=1
  
  return ct,ep,err,optx1,optx2,optf
```

```python
#initial point used
x1=0  
x2=0
#step size
lr=0.1
ct,ep,err,optx1,optx2,optf=gradplot(x1, x2, lr, 1)
#print(x1, x2)
ct,ep,err,optx1,optx2,optf=gradplot(x1, x2, lr, 1, optf)
plt.plot(ep, err)
plt.show
print('no. of iterations=', ct,'\noptimal x1=',str(optx1),"\noptimal x2=",str(optx2),"\noptimal f=",str(optf))
```

Inference: The error goes down with the number of iterations and then the curve flattens ie. the speed of going down decreases


```python
lr=0.05
x1=0
x2=0
iter=[]
lrs=[]
while(lr<1):
  print(lr)

  ct,ep,err,optx1,optx2,optf=gradplot(x1, x2, lr, 1)
  ct,ep,err,optx1,optx2,optf=gradplot(x1, x2, lr, 1, optf)
  iter.append(ct)
  lrs.append(lr)
  lr+=0.05
  

plt.plot(lrs, iter)
plt.show
```

Inference: There is an optimal step size which converges the graph in the least number of iterations.

```python
lr=0.5
x1=0
x2=0
iter=[]
lrs=[]
g=0.5
while(g<1):
  print(g)
  ct,ep,err,optx1,optx2,optf=gradplot(x1, x2, lr, g)
  ct,ep,err,optx1,optx2,optf=gradplot(x1, x2, lr, g, optf)
  iter.append(ct)
  lrs.append(g)
  g+=0.1

plt.plot(lrs, iter)
plt.show
```

Inference: With an increasing value of gamma, lesser no. of iterations are required because the absolute value of gradient with respect to x2 increases hence with a similar learning rate, the difference between ith and (i+1)th iteration is huge.  


```python
ct,ep,err,optx1,optx2,optf=gradplot(0, 0, 0.1, -1)
print(ct)
plt.plot(ep, err,'.')
plt.show
```

Since we have a saddle point, the graph falls off the minima(as assumed by gradient descent) and the error plunges. Also, it givess the mistaken minima or won't give any point at all if the learning rate is high enough

```python
def f1(x1, x2):
  return np.exp(x1+3*x2)

def f2(x1, x2):
  return np.exp(x1-3*x2)

def newton(x1, x2, lr=0.1):
  prev=1000000
  cur=1
  ct=1
  ep=[]
  err=[]

  while (1):

    dfdx1=f1(x1,x2)+f2(x1,x2)-np.exp(-x1)
    dfdx2=3*f1(x1,x2)-3*f2(x1,x2)

    d2fdx12=f1(x1,x2)+f2(x1,x2)+np.exp(-x1)
    d2fdx22=9*f1(x1,x2)+9*f2(x1,x2)
    #d2fdx1dx2=3*f1(x1,x2)-3*f2(x1,x2)

    #hessian= np.array([[d2fdx12,d2fdx1dx2],[d2fdx1dx2,d2fdx22]]) 
    #hessinv= np.linalg.inv(hessian)


    x1=x1-lr*dfdx1/d2fdx12
    x2=x2-lr*dfdx2/d2fdx22
    
    cur=(f1(x1,x2)+f2(x1,x2)+np.exp(-x1))*np.exp(-0.1)
    ep.append(ct)
    err.append(abs(2.55927-cur))
    if(cur==prev):
      optx1=x1
      optx2=x2
      optf=cur
      break
    prev=cur
    ct+=1
  
  return ct,ep,err,optx1,optx2,optf
```

```python
ct,ep,err,optx1,optx2,optf=newton(2,1)
plt.plot(ep, err)
plt.show
print('no. of iterations=', ct,'\noptimal x1=',str(optx1),"\noptimal x2=",str(optx2),"\noptimal f=",str(optf))
```

```python
ct,ep,err,optx1,optx2,optf=newton(-2,1)
plt.plot(ep, err)
plt.show
print('no. of iterations=', ct,'\noptimal x1=',str(optx1),"\noptimal x2=",str(optx2),"\noptimal f=",str(optf))
```
